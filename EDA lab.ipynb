```python
# prompt: draw a plot that shows how many people left vs how many people stay group by their num_project

import matplotlib.pyplot as plt

# Group by 'number_project' and 'left' and count the number of employees
project_left_counts = df.groupby(['number_project', 'left'])['left'].count().unstack()

# Plot the counts
project_left_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Number of Employees Left vs. Stayed by Number of Projects')
plt.xlabel('Number of Projects')
plt.ylabel('Number of Employees')
plt.legend(['Stayed', 'Left'])
plt.show()
```

```python
# prompt: draw a plot that shows the effect of salary on left

# Group by 'salary' and 'left' and count the number of employees
salary_left_counts = df.groupby(['salary', 'left'])['left'].count().unstack()

# Plot the counts
salary_left_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Number of Employees Left vs. Stayed by Salary')
plt.xlabel('Salary Level')
plt.ylabel('Number of Employees')
plt.legend(['Stayed', 'Left'])
plt.show()
```

```python
# prompt: draw a plot that shows attrition group by role

# Group by 'sales' and 'left' and count the number of employees
role_left_counts = df.groupby(['role', 'left'])['left'].count().unstack()

# Plot the counts
role_left_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Number of Employees Left vs. Stayed by Role')
plt.xlabel('Role')
plt.ylabel('Number of Employees')
plt.legend(['Stayed', 'Left'])
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
# prompt: draw a plot that shows the left based on their satisfaction level using histogram

plt.figure(figsize=(10, 6))
plt.hist(df[df['left'] == 1]['satisfaction_level'], bins=20, alpha=0.5, label='Left')
plt.hist(df[df['left'] == 0]['satisfaction_level'], bins=20, alpha=0.5, label='Stayed')
plt.xlabel('Satisfaction Level')
plt.ylabel('Number of Employees')
plt.title('Histogram of Satisfaction Level for Employees who Left vs. Stayed')
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/theleadio/datascience_demo/master/car_sales.csv")
df
```

business question: know the selling price and factors that affect the price

```python
corr = df.corr(numeric_only=True)
corr
```

```python

plt.figure(figsize=(12,10))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
# prompt: draw a heatmap using seaborn for corr

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# prompt: visualise the relationship between car manufacture and price

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.boxplot(x='Manufacturer', y='Price_in_thousands', data=df)
plt.title('Relationship Between Car Manufacturer and Price')
plt.xlabel('Manufacturer')
plt.ylabel('Price')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# prompt: visualise the relationship between car manufacturer and price_in_thousands, turn the label 90 degrees, using a vibrant bright palette, draw the chart in HD, hide the future warnigns

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16, 10), dpi=300)  # Increased figure size and DPI for HD
sns.boxplot(x='Manufacturer', y='Price_in_thousands', data=df, palette='bright')
plt.title('Relationship Between Car Manufacturer and Price')
plt.xlabel('Manufacturer')
plt.ylabel('Price (in thousands)')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()  # Ensures labels don't overlap
plt.show()
```

The illustration shows the relationship between various car manufacturers and their prices, displayed in thousands of dollars.

- X-axis (Manufacturer): Different car brands such as Acura, Audi, BMW
- Y-axis (Prices in Thousands): The price of the cars from each manufacturer, shown in thousands of dollars.
- Box-plots: Each box plot represents the distribution of prices for a specific manufacturer, showing the median, quartiles, and potential outliers.
    - The line inside each box represents the median price for that manufacturer.
    - The end of the box represents the 25th and 75th precentiles.
    - The 'whiskers' extend to the smallest and largest values, excluding outliers
    - Any outliers (points outside the typical range) are marked as circles.

  These visualisation helps compares the price ranges of different car brands. For instance, manufacturers like Mercedes-Benz, Porsche, and Lexus tend to have higher price ranges compared to brands like Saturn, Volkwagens, and Hyundai. Outliers for brands like Chevrolet, Cadillac, and Subaru show visually high or low prices for certain models.
