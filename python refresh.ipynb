```python
# prompt: write a list with 10 different fruits

fruits = ["apple", "banana", "orange", "grape", "watermelon", "strawberry", "blueberry", "raspberry", "mango", "pineapple"]
print(fruits)
```

```python
# prompt: write a list comprehension to capitalise fruits

capitalized_fruits = [fruit.capitalize() for fruit in fruits]
print(capitalized_fruits)
```

```python
# prompt: write a for loop to capitalise fruits and store in a new list call upper_fruits

upper_fruits = []
for fruit in fruits:
  upper_fruits.append(fruit.upper())
print(upper_fruits)
```

```python
# prompt: create a list with 1 to 10

my_list = list(range(1, 11))
print(my_list)
```

```python
# prompt: filter the numbers with only even numbers

even_numbers = [num for num in my_list if num % 2 == 0]
print(even_numbers)
```

```python
# add a condotion at the back
# add a transformation in front
even_numbers = [n^2 for n in my_list if n % 2 == 0]
even_numbers
```
